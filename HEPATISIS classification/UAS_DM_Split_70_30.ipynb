## Pre-processing

```python
#mengganti data yang bernilai ? dengan NaN untuk mengetahui cell yang kosong
df.replace("?", np.nan, inplace=True)
df.isnull().sum()
```

```python
#isi cell yang kosong dengan nilai tengah tiap kolom/atribut
df.fillna(df.median(), inplace=True)
df.isnull().sum()
```

```python
#hapus data duplikat
df.drop_duplicates()
```

```python
df.info()
```

```python
#rubah format data
df["STEROID"] = df["STEROID"].astype('int64')
df["FATIGUE"] = df["FATIGUE"].astype('int64')
df["MALAISE"] = df["MALAISE"].astype('int64')
df["ANOREXIA"] = df["ANOREXIA"].astype('int64')
df["LVER_BIG"] = df["LVER_BIG"].astype('int64')
df["LIVER_FIRM"] = df["LIVER_FIRM"].astype('int64')
df["SPLEEN_PALPABE"] = df["SPLEEN_PALPABE"].astype('int64')
df["SPIDERS"] = df["SPIDERS"].astype('int64')
df["ASCITIES"] = df["ASCITIES"].astype('int64')
df["VARICES"] = df["VARICES"].astype('int64')
df["BILIRUBIN"] = df["BILIRUBIN"].astype('float')
df["ALK_PHOSPATE"] = df["ALK_PHOSPATE"].astype('int64')
df["SGOT"] = df["SGOT"].astype('int64')
df["ALBUMIN"] = df["ALBUMIN"].astype('float')
df["PROTIME"] = df["PROTIME"].astype('int64')

df.info()
```

## Visualize data

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, fmt='.1g', cmap="Blues_r", cbar=False, linewidths=0.5, linecolor='grey');
```

## Spliting Data

```python
X = df.iloc[:,1:19]
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
len(X_train), len(X_test)
```

## normalisasi data

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Permodelan

## KNN K3

```python
knn3 = KNeighborsClassifier(n_neighbors=3)
knn3.fit(X_train, y_train)
```

```python
KNN3_score = knn3.score(X_test, y_test)
print("KNN 3 Classifier Accuracy :", KNN3_score*100)
```

```python
y_pred_knn3 = knn3.predict(X_test)
cf_matrix_knn3 = confusion_matrix(y_test, y_pred_knn3)
sns.heatmap(cf_matrix_knn3, annot=True)
plt.title("Confusion Matrix for KNN 5 Classifier", fontsize=15);
```

```python
print(classification_report(y_test, y_pred_knn3))
```

## KNN K5

```python
knn5 = KNeighborsClassifier()
knn5.fit(X_train, y_train)
```

```python
KNN5_score = knn5.score(X_test, y_test)
print("KNN 5 Classifier Accuracy :", KNN5_score*100)
```

```python
y_pred_knn5 = knn5.predict(X_test)
cf_matrix_knn5 = confusion_matrix(y_test, y_pred_knn5)
sns.heatmap(cf_matrix_knn5, annot=True)
plt.title("Confusion Matrix for KNN 5 Classifier", fontsize=15);
```

```python
print(classification_report(y_test, y_pred_knn5))
```

## KNN K7

```python
knn7 = KNeighborsClassifier(n_neighbors=7)
knn7.fit(X_train, y_train)
```

```python
KNN7_score = knn7.score(X_test, y_test)
print("KNN 7 Classifier Accuracy :", KNN7_score*100)
```

```python
y_pred_knn7 = knn7.predict(X_test)
cf_matrix_knn7 = confusion_matrix(y_test, y_pred_knn7)
sns.heatmap(cf_matrix_knn7, annot=True)
plt.title("Confusion Matrix for KNN 7 Classifier", fontsize=15);
```

```python
print(classification_report(y_test, y_pred_knn7))
```
